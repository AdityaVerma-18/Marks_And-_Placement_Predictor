```python
# logistic regression is perfecct model for this data because they are easily seperable
# seperate input and output columns
# independent vareiable
x=df.iloc[:,0:2]
x.head()
```

```python
# dependent variable
y=df.iloc[:,[2]]
y.head(2)
```

```python
# train test split
from sklearn.model_selection import train_test_split
train_test_split(x,y,test_size=0.1)
# 0.1 sognifies 10 per of data will be used for testing the model
```

```python
print("numpy", np .__version__)
print("pandas", pd .__version__)
print("sklearn",sklearn .__version__)
```

```python
# storing data in different variabales
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.1)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
ytrain.head()
```

```python
ytest.head()
```

```python
# scaling
from sklearn.preprocessing import StandardScaler
```

```python
Scaler = StandardScaler()
xtrain=Scaler.fit_transform(xtrain)
```

```python
xtrain
```

```python
xtest=Scaler.transform(xtest)
xtest
```

```python
ytrain
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(xtrain,ytrain)
```

```python
clf.predict(xtest)
```

```python
ypred=clf.predict(xtest)
ytest
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,ypred)
```
